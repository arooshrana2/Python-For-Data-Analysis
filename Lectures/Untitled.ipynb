```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'House':['Taxila', 'Taxila', 'Ujjain', 'Nalanda', 'Ujjain', 'Nalanda', 'Ujjain', 'Taxila', 'Nalanda', 'Nalanda'],
                  'Score':[10,20,12,15,20,14,18,20,10,12]})
```

```python
df.sort_values(['House', 'Score'], ascending=[True, False])
```

```python
df.groupby('House')['Score'].nlargest(2).reset_index().drop('level_1', axis=1).groupby('House').agg('sum').reset_index()
```
