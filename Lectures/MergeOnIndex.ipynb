```python
import pandas as pd
import numpy as np
```

```python
df1_hr = pd.DataFrame({'key1':['LA', 'LA', 'SF', 'SF', 'LA'],
                      'key2':[1,2,2,1,1],
                      'data':[34,45,56,76,54]})
```

```python
df2_hr = pd.DataFrame(np.arange(10).reshape(5,2),
                     index=[['LA','LA','LA','SF','SF'],
                           [2,2,1,2,1]],
                     columns=['one', 'two'])
```

```python
df3_hr = pd.DataFrame(np.arange(10).reshape(5,2),
                     index=[[2,2,1,2,1],
                            ['SF','SF','SF','LA','LA']],
                     columns=['one', 'two'])
```

```python
df1_hr
```

```python
df2_hr
```

```python
pd.merge(df1_hr, df2_hr, left_on=['key1', 'key2'], right_index=True)
```

```python
df3_hr
```

```python
pd.merge(df1_hr, df3_hr, left_on=['key2','key1'], right_index=True)
```
